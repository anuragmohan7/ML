```python
# we should drop Irrelevant, because it is not needed, so we drop irrelevant rows
df.drop(df.index[(df['target']=='Irrelevant')],axis=0,inplace=True)
df
```

```python
#reset index
df.reset_index(drop=True,inplace=True)
df
```

```python
#remove unwanted column
df.drop(['id','social'],axis=1,inplace=True)
df
```

```python
#now target column convert positive--> 1, negative --> -1, neutral --> 0
df['target'] = df['target'].map({'Positive': 1, 'Negative': -1,'Neutral':0})
df
```

```python
df.dtypes
```

```python
# now we apply nlp to text column
tweets=df.text
tweets
```

```python
# tokenization
from nltk.tokenize import TweetTokenizer  #TweetTokenizer used for dataframe
tk=TweetTokenizer()
tweets=tweets.apply(lambda x:tk.tokenize(x)).apply(lambda x:" ".join(x))  #1st lambda function used for tokenize
                                                                            #2nd lambda function used separating tokens with spaces.
tweets
```

```python
# remove special characters
tweets=tweets.replace('[^a-zA-Z0-9]',' ',regex=True) #replacing not [^a-zA-Z0-9] with space,  regex-- regular expression
tweets
```

```python
# not important step
# to get a meaningful words we remove words which have lessthan 3 letters
tweets = tweets.apply(lambda x: ' '.join([i for i in tk.tokenize(x) if len(i) >= 3]))
tweets
```

```python
# stemming --> used to get root word
from nltk.stem import SnowballStemmer
ss=SnowballStemmer('english')
tweets=tweets.apply(lambda x:[ss.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x: ' '.join(x))
tweets
```

```python
# identification of stop word
from nltk.corpus import stopwords
sw=stopwords.words('english')
```

```python
#remove stopwords from tweets
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda x:' '.join(x))
tweets
```

```python
# vectorization text--->number
# we use tfidf --> term frequency inverse document frequency

from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
data=vec.fit_transform(tweets)  # now vectorized tweets are stored in data
print(data)
```

```python
y=df['target'].values
```

```python
# now data ==> input
#     y ==> output
```

```python
# now we apply machine learning classification for prediction
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.30,random_state=1)
```

```python
#multiple model creation 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
knn=KNeighborsClassifier(n_neighbors=7)
nai=BernoulliNB()
sup=SVC()
```

```python
lst=[knn,nai,sup]
for i in lst:
    print("Model name is ---> ",i)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("Performance Evaluation of ",i,"****************")

    print('accuracy_score === ',accuracy_score(y_test,y_pred))
    print('-'*100)
```
