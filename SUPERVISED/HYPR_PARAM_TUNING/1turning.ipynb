```python
df['trestbps']=df['trestbps'].fillna(df['trestbps'].mode()[0])
df['restecg']=df['restecg'].fillna(df['restecg'].mode()[0])
df['thalach']=df['thalach'].fillna(df['thalach'].mean())
df['ca']=df['ca'].fillna(df['ca'].mode()[0])
df['thal']=df['thal'].fillna(df['thal'].mode()[0])
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
before_score=accuracy_score(y_test,y_pred)
before_score
```

```python
#HYPER PARAMETER TUNNING

#GridSearchCV
from sklearn.model_selection import GridSearchCV
model1=KNeighborsClassifier()
parameter={'n_neighbors':[3,5,7,9],'weights':['uniform','distanace']}
grid=GridSearchCV(model1,parameter,cv=10,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
# to print best parameter
print(grid.best_params_)
```

```python
model2=KNeighborsClassifier(n_neighbors=3,weights='uniform')
model2.fit(x_train,y_train)
y_pred1=model2.predict(x_test)
after_score=accuracy_score(y_test,y_pred1)
after_score
```
