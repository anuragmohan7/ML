```python
from imblearn.under_sampling import RandomUnderSampler
rus =RandomUnderSampler(random_state=42)

x=df.drop(['Preference'],axis=1)
y=df['Preference']

x_under,y_under=rus.fit_resample(x,y)
df=x_under.copy()
df['Preference']=y_under
```

```python
df['Preference'].value_counts()

```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])
df['Education_Level']=lab.fit_transform(df['Education_Level'])
df['Preferred_Activities']=lab.fit_transform(df['Preferred_Activities'])
df['Location']=lab.fit_transform(df['Location'])
df['Favorite_Season']=lab.fit_transform(df['Favorite_Season'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=7)
nai=BernoulliNB()
sup=SVC()
lst=[knn,nai,sup]

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay

for i in lst:
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("Accuracy score of ",i)
    score=accuracy_score(y_test,y_pred)
    print(score)
    print("confusion_matrix of ",i)
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
    print("report of ",i)
    report=classification_report(y_test,y_pred)
    print(report)
    print('-'*100)
```
