```python
from imblearn.over_sampling import RandomOverSampler
rus=RandomOverSampler(random_state=1)

xx=df.drop(['salary'],axis=1)
yy=df['salary']

x_over,y_over=rus.fit_resample(xx,yy)
df=x_over.copy()
df['salary']=y_over
```

```python
df['salary'].value_counts()
```

```python
df.drop(['race','sex'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['workclass']=lab.fit_transform(df['workclass'])
df['education']=lab.fit_transform(df['education'])
df['marital-status']=lab.fit_transform(df['marital-status'])
df['occupation']=lab.fit_transform(df['occupation'])
df['relationship']=lab.fit_transform(df['relationship'])
df['native-country']=lab.fit_transform(df['native-country'])
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=7)
nai=BernoulliNB()
sup=SVC()
lst=[knn,nai,sup]

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay

for i in lst:
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("Accuracy score of ",i)
    score=accuracy_score(y_test,y_pred)
    print(score)
    print("confusion_matrix of ",i)
    cm=confusion_matrix(y_test,y_pred)
    print(cm)
    print("report of ",i)
    report=classification_report(y_test,y_pred)
    print(report)
    print('-'*100)
    
```
