```python
numeric_columns=df.select_dtypes(include='number').columns
numeric_columns
```

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers



for col in numeric_columns:
    outliers = detect_outliers_iqr(df[col])
    print(f'Number of outliers in {col}: {len(outliers)}')
```

```python
outlier_columns=['Temperature_C','Vibration_mms','Sound_dB','Oil_Level_pct','Power_Consumption_kW','Maintenance_History_Count','Failure_History_Count','Error_Codes_Last_30_Days','AI_Override_Events']  #these are the columns which have outliers
def remove_outliers(df,columns):
    for col in columns:
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        iqr=q3-q1

        lower_bound=q1-1.5*iqr
        upper_bound=q3+1.5*iqr

        df=df[(df[col]>lower_bound)&(df[col]<upper_bound)]
    return df
df_cleaned=remove_outliers(df,outlier_columns)      # df_cleaned is outlier free dataframe
```
