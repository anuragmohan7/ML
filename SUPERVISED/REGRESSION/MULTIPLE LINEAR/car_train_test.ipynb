```python
#Next step is we should encode the object columns, The column name is a big column So one hot encoding may Result in overfitting, 
#   Here we can use label Encoding But Label engoding have drawback of hierarchy formation
#   So here we use a different approach, In the name column Each brand have Different model cars, So here we only extract the brand

#this step is used to reduce the unique value
```

```python
df.shape
```

```python
df['Name'].unique()
len(df['Name'].unique())
```

```python
#Extracting the brand from name, By adding a new column brand
df['brand']=df['Name'].apply(lambda x:x.split()[0]) #Apply function is used in dataframe If we use lambda function
```

```python
df.shape
```

```python
df['brand'].unique()
len(df['brand'].unique())
```

```python
df['brand'].value_counts()
```

```python
#Now we can handle the data Because the brand column contains only Less number of values
```

```python
#So the encoding columns are Location, fuel-type, Transmission, ownertype, brand
#We cannot use label Encoding (Because of hierarchy formation) and one hot encoding (Large number of columns) here,
#So we use Another encoding called get_dummies

#we can drop unnamed,name
```

```python
#Get dummies are inside pandas
dfe=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','brand']],drop_first=True,dtype=int)
dfe #now these data are encoded
```

```python
dfe.shape
```

```python
#now we should concatnate Encoded data with df data
```

```python
df1=pd.concat([df,dfe],axis=1)
df1
```

```python
df1.shape
```

```python
df1.columns
```

```python
#next step is dropping unwanted columns unnamed, name, location(Already encoded So we don't need this column), fuel-type, transmission, owner type, new price(large missing value), brand
df1.drop(['Unnamed: 0', 'Name', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'New_Price', 'brand'], inplace=True,axis=1)
df1
```

```python
#next step is remove unit from mileage,engine,power
df1['Mileage']=df1['Mileage'].str.replace('km/kg','') # '' is used to replace with non-space
df1['Mileage']=df1['Mileage'].str.replace('kmpl','')
df1['Engine']=df1['Engine'].str.replace('CC','')
df1['Power']=df1['Power'].str.replace('bhp','')

df1
```

```python
# We should replace string null value from mileage engine power to 0
df1['Mileage']=df1['Mileage'].str.replace('null','0')
df1['Power']=df1['Power'].str.replace('null','0')
df1['Engine']=df1['Engine'].str.replace('null','0')
df1



```

```python
#Next step is to convert object data type of columns mileage engine power to int
```

```python
dfe.dtypes
```

```python
df1['Mileage']=df1['Mileage'].astype(float)
df1['Engine']=df1['Engine'].astype(float)
df1['Power']=df1['Power'].astype(float)
df1.dtypes
```

```python
df1.isna().sum()
```

```python
#Replace numerical value 0 to nan(missing value, so that we can fillna later)
df1.loc[df1.Mileage==0,'Mileage']=np.nan #(nan means missing value)
df1.loc[df1.Engine==0,'Engine']=np.nan
df1.loc[df1.Power==0,'Power']=np.nan
```

```python
df1.isna().sum()
```

```python
#Next step is filling missing value in nan place
df1['Mileage'].unique()
```

```python
df1['Mileage']=df1['Mileage'].fillna(df1['Mileage'].mean())
df1['Engine']=df1['Engine'].fillna(df1['Engine'].mean())
df1['Power']=df1['Power'].fillna(df1['Power'].mean())
df1['Seats']=df1['Seats'].fillna(df1['Seats'].mode()[0])
```

```python
df1.isna().sum()
```
