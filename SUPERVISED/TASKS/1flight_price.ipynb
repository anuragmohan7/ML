```python
df1[['day', 'month', 'year']] = df1['Date_of_Journey'].str.split('/', expand=True)

df1['Hours'] = df1['Duration'].str.extract(r'(\d+)h')
df1['Minutes'] = df1['Duration'].str.extract(r'(\d+)m')
df1
```

```python
dfe=pd.get_dummies(df1[['Route']],drop_first=True,dtype=int)
dfe
```

```python
df2=pd.concat([df1,dfe],axis=1)
df2
```

```python
df2.drop(['Route'],axis=1,inplace=True)
df2
```

```python
df2.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df2['Airline']=lab.fit_transform(df2['Airline'])
df2['Source']=lab.fit_transform(df2['Source'])
df2['Destination']=lab.fit_transform(df2['Destination'])
df2['Dep_Time']=lab.fit_transform(df2['Dep_Time'])
df2['Total_Stops']=lab.fit_transform(df2['Total_Stops'])
df2['Additional_Info']=lab.fit_transform(df2['Additional_Info'])
df2['day']=lab.fit_transform(df2['day'])
df2['month']=lab.fit_transform(df2['month'])
df2['year']=lab.fit_transform(df2['year'])
df2['Hours']=lab.fit_transform(df2['Hours'])
df2['Minutes']=lab.fit_transform(df2['Minutes'])
df2['Arrival']=lab.fit_transform(df2['Arrival'])
```

```python
df2.columns
x=df2.drop(['Price'],axis=1)
x
```

```python
y=df2['Price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('mean_absolute_percentage_error = ',mean_absolute_percentage_error(y_test,y_pred))
print('r2_score = ',r2_score(y_test,y_pred))
```
