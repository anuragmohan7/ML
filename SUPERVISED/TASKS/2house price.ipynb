```python
outlier_columns=['price','sqft_lot']  
def remove_outliers(df,columns):
    for col in columns:
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        iqr=q3-q1

        lower_bound=q1-1.5*iqr
        upper_bound=q3+1.5*iqr

        df=df[(df[col]>lower_bound)&(df[col]<upper_bound)]
    return df
df_cleaned=remove_outliers(df,outlier_columns)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_cleaned.corr(numeric_only=True),annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5) #annot means write values inside cell,fmt .2f means Floating number with two decimal places
plt.title('heatmap')
plt.show()
```

```python
df_numeric=df.drop(['sqft_lot','yr_renovated','day','STATEZIP'],axis=1)
df_numeric
```

```python
df_encode=pd.get_dummies(df_numeric[['street','city']],drop_first=True,dtype=int)
df_encode
```

```python
dfe=pd.concat([df_numeric,df_encode],axis=1)
dfe
```

```python
dfe.drop(['street','city'],axis=1,inplace=True)
dfe
```

```python
y=dfe.iloc[:,0]
x=dfe.iloc[:,1:]
```

```python
dfe.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print('before')
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('mean_absolute_percentage_error = ',mean_absolute_percentage_error(y_test,y_pred))
print('r2_score = ',r2_score(y_test,y_pred))
```

```python
#HYPER PARAMETER TUNNING

#GridSearchCV
from sklearn.model_selection import GridSearchCV
model1=DecisionTreeRegressor()
parameter={'criterion':['squared_error','friedman_mse'],'splitter':['best','random']}
grid=GridSearchCV(model1,parameter,cv=10,scoring='r2')
grid.fit(x_train,y_train)


```

```python
print(grid.best_params_)
```

```python
model2=DecisionTreeRegressor(criterion='squared_error',splitter='random')
model2.fit(x_train,y_train)
y_pred2=model2.predict(x_test)
print('after')
print('mean_absolute_percentage_error = ',mean_absolute_percentage_error(y_test,y_pred2))
print('r2_score = ',r2_score(y_test,y_pred2))

```
