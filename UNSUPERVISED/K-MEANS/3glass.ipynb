```python
from sklearn.cluster import KMeans
lst=[]
for i in range(1,11):
    data=KMeans(n_clusters=i,init='k-means++',random_state=42)
    data.fit(df)
    lst.append(data.inertia_)

```

```python
plt.plot(range(1,11),lst,marker='o')
plt.xticks(range(1,11))
plt.plot
```

```python
data2=KMeans(n_clusters=7,init='k-means++',random_state=42)
data2.fit(df)
y_pred=data2.predict(df)
y_pred
```

```python
df['clusters']=y_pred
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
knn=KNeighborsClassifier(n_neighbors=7)
nai=GaussianNB()
svm=SVC()
lst2=[knn,nai,svm]
```

```python
for i in lst2:
    print("Model name is ---> ",i)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print("Performance Evaluation of ",i,"****************")
    print('accuracy_score === ',accuracy_score(y_test,y_pred))
    print('-'*100)
```
