```python
#how to draw a line
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[7,5,4,8,12,3,2]
plt.plot(x,y)
```

```python
x=[12,78,45,3,33,22,8]
y=[12,2,4,5,7,6,4]
plt.plot(x,y)
```

```python
#using array
import numpy as np
a=np.array([1,2,3,4,5,6,7])
b=np.array([4,7,5,1,2,6,3])
plt.plot(a,b)
```

```python
#only points
c=[1,2,7,8,9,6]
plt.plot(c,'*')
```

```python
plt.plot(c,marker='o')
```

```python
#       * ==> marker,  : ==> dotted line,  g ==> color
plt.plot(c,'*:g') #other marker ==> o,+,s,h
                  #other line ==> :,--,-
```

```python
#marker properties
plt.plot(c,'-o',ms=15,mec='y',mfc='r') #ms ==> marker size, mec ==> marker edge color, mfc ==> marker face color
```

```python
#line properties
plt.plot(c,'o:g',linewidth=7)
```

```python
plt.plot(c,'o--b',mfc='r',mec='r')
```

```python
#multiple line in graph
x1=[4,5,7,8]
y1=[6,5,8,4]
x2=[7,5,3,2]
y2=[9,6,5,4]
plt.plot(x1,y1,'o:')
plt.plot(x2,y2,'o:')
```

```python
# x,y-axis   labels
height=[155,145,165,170]
weight=[60,55,70,80]
plt.plot(height,weight,'o')
plt.xlabel("height")
plt.ylabel("weight")
plt.title("height - weight graph")
```

```python
#velocity-displacement graph
velo=[45,78,62,63]
disp=[44,22,56,77]
plt.plot(velo,disp,'o:',mfc='r',mec='r')
plt.xlabel("velocity",size=15)
plt.ylabel("displacement",size=15)
plt.title("velocity-displacement graph",size=20,color='r')

#grid
plt.grid()  #axis='x' or 'y',  linestyle='dashed'
```

```python
#subplot ==> used for graph partician
x1=[1,2,3,4]
y1=[3,4,7,9]
plt.subplot(1,2,1) #1,2 --> one row two columns, 1 --> graph 1
plt.plot(x1,y1,'o:')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph 1")
plt.suptitle("graph 1 - graph 2") #main heading for both graph

x2=[4,3,2,1]
y2=[3,6,8,7]
plt.subplot(1,2,2) #1,2 --> one row two columns, 2 --> graph 2
plt.plot(x2,y2,'o:r',mfc='r',mec='r')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph 2")
```
