```python
import matplotlib.pyplot as plt
```

```python
#scatter plot  ==> only represent in points
x1=[1,2,3,4]
y1=[3,4,8,7]
plt.scatter(x1,y1)

x2=[4,5,7,8]
y2=[7,4,1,5]
plt.scatter(x2,y2)
```

```python
#barchart (it is suitable when count based graph)
x=[4,5,7,8,9]
y=[5,6,3,2,4]
plt.bar(x,y)
```

```python
country=['india','us','aus','america','cuba']
winners=[4,7,8,6,5]
plt.bar(country,winners,color='r')
plt.xlabel('country')
plt.ylabel('winner count')
plt.title('country - winners')
```

```python
country=['india','us','aus','america','cuba']
winners=[4,7,8,6,5]
plt.barh(country,winners,color='r')
plt.ylabel('country')
plt.xlabel('winner count')
plt.title('country - winners')
```

```python
#pie-chart (works based on percentage, only one axis required)
data=[10,20,30,40]
lbl=['apple','mango','grape','pineapple']
plt.pie(data,labels=lbl,autopct='%0.1f%%')
plt.legend(loc='upper left') #used to define colors
plt.tight_layout() #reduce overlap
plt.plot
```

```python
#histogram (it number of frequency observation, only need one axis) 
x=[1,1,1,2,2,3,3,3,3,3,4,4,4,5,5,5,5,5,6,6,6,6]
plt.hist(x)
plt.xticks([1,2,3,4,5,6])
plt.grid(axis='y')
plt.plot()
```
