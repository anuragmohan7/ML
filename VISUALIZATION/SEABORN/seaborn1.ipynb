```python
import seaborn as sns
df=sns.load_dataset('tips') #tips is built in dataset inside seaborn for study purpose 
df
```

```python
df.isna().sum()
```

```python
df.dtypes #category === object
```

```python
# FEATURE SELECTION

#steps...
# 1. CO-RELATION --> it is used to check how numeric columns related to other numeric columns
# 2. CHI-SQUARE TEST ---> it is used to check how object columns related to other object columns
```

```python
#co-relation
df.corr(numeric_only=True)
```

```python
# HEAT-MAP   --->  graphical plot of co-relation
sns.heatmap(df.corr(numeric_only=True),fmt='2f',annot=True,cmap='coolwarm')
```

```python
# JOINT PLOT  --->   just like scatter plot also it can plot regression line or histogram  withit (only numeric columns)
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') #kind is used to plot regression line withit (kind='reg','hex')
```

```python
sns.jointplot(x='tip',y='size',data=df,kind='reg')
```

```python
sns.jointplot(x='size',y='total_bill',data=df,kind='hex')
```

```python
# PAIR PLOT --->  it is used to plot all joint plot in a single plot (only numeric column)
sns.pairplot(df,kind='reg')
```

```python
sns.pairplot(df,hue='time')  # hue is used to plot pair plot corresponding to an object column
```

```python
sns.pairplot(df,hue='smoker')
```
